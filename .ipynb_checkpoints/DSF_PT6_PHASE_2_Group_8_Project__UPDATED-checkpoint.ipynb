```python
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

#### <span style='color:Green'>SIMPLE LINEAR REGRESSION</span>


```python
columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'yr_built', 'condition_coded']

# Calculate the number of rows and columns for the subplot grid
num_cols = 3
num_rows = (len(columns) + num_cols - 1) // num_cols

# Create subplots for each column
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10), sharey=True)

# Flatten the axes array if necessary
axes = axes.flatten()

# Perform OLS regression for each column and plot the regression line
for i, column in enumerate(columns):
    # Extract the features and target
    X = df[[column]]
    y = df['price']

    # Add constant term to the features
    X = sm.add_constant(X)

    # Fit OLS regression model
    model = sm.OLS(y, X).fit()

    # Plot the scatter plot
    sns.scatterplot(x=column, y='price', data=df, ax=axes[i], alpha=0.5)

    # Plot the regression line
    sns.lineplot(x=X[column], y=model.predict(X), ax=axes[i], color='red')

    axes[i].set_title(f'{column.capitalize()} vs Price')
    axes[i].set_ylabel('Price')
    axes[i].set_xlabel(column.capitalize())

# Hide any empty subplots
for ax in axes[len(columns):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

Sqft_living is highly correlated to house prices .The bigger the house the higher the cost <br>
Sqft_lot does not have a great impact on house pricing <br>

#### <span style='color:Blue'>Summary of a Linear regression table </span>
 Dependent Variable (MEDV):

This is the variable we are trying to predict called median value. In this case Price of the House
Model (Ordinary Least Squares)

This is the method used for estimating the coefficients of the linear regression model.
Coefficients:<br>

Represents the estimated value of the dependent variable when all independent variables are zero.

RM, CRIM, LSTAT:-For every one unit of change in the independent variable, there is a change in the dependent variable <br>

Coefficient Statistics:

coef (Coefficient Estimate): This is  estimated value of the coefficient.<br>
std err (Standard Error):<br>
 the precision of the estimate. The smaller the values, the better <br>
t (t-statistic): <br>
Tests the null hypothesis that the coefficient is equal to zero. Larger absolute values indicate more significance.<br>
P>|t| (p-value): The p-value associated with the t-statistic. A small p-value (< 0.05) suggests that the variable is statistically significant.<br>
[0.025 0.975] (Confidence Interval): The 95% confidence interval for the coefficient. If it contains zero, the variable may not be significant.<br>
 Model Fit:

R-squared: Represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s). A higher R-squared indicates a better fit.<br>
Adj. R-squared: Similar to R-squared but adjusted for the number of predictors. It penalizes excessive use of variables.<br>
F-statistic-This is the overall significance of the model .It assesses whether one independent variable is significant in explaining the dependent variable .A small p-value( <0.05)indicates that the model is significant <br>

Skewness:

Meaning: Skewness measures the asymmetry of the distribution of residuals. It indicates whether the residuals are symmetrically distributed around the mean.
 Kurtosis:

Meaning: Kurtosis measures the “tailedness” of the distribution of residuals. It indicates the heaviness of the tails relative to a normal distribution.








##Individual Simple Linear Regression

```python
#Simple Regression for sqft_living
X = df['sqft_living']
y = df['price']

X = sm.add_constant(X)

# Fit OLS regression model
sqft_living_model = sm.OLS(y, X).fit()

# Plotting the scatter plot and regression line
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sqft_living', y='price', data=df, alpha=0.5)
sns.lineplot(x=X['sqft_living'], y=sqft_living_model.predict(X), color='red')

plt.title('sqft_living vs Price')
plt.xlabel('sqft_living')
plt.ylabel('Price')

plt.show()

# Print the summary of the regression results
print(sqft_living_model.summary())
print('')
print("f_value:",sqft_living_model.fvalue)
print( "p_value:",sqft_living_model.f_pvalue)
```

##Modified Models using ridge regression

##Why ridge regression:
Compared to other methods of handling multicollinearity, such as variable selection, principal component analysis (PCA), and Lasso regression, ridge regression has several advantages:

1. Preservation of Information: Ridge regression preserves all the variables in the model, unlike variable selection methods that discard some variables.

2. Interpretability: Ridge regression retains the interpretability of the original variables, as it shrinks rather than eliminates coefficients.

3. Control over Shrinkage: Ridge regression allows for controlled shrinkage of coefficients through the regularization parameter (lambda or alpha).

4. Less Sensitive to Outliers: Ridge regression is less sensitive to outliers compared to methods like PCA, which can be heavily influenced by extreme observations due to the use of covariance matrix.

```python
#sqft_living
# Given correlation data
# Convert correlation data to a DataFrame
correlation = pd.DataFrame(correlation)

# Prepare data for simple linear regression
X_simple = correlation['sqft_living'].values.reshape(-1, 1)  # Independent variable for simple regression
y = correlation['price']  # Dependent variable

# Add constant term to independent variable for simple linear regression
X_simple_const = sm.add_constant(X_simple)

# Fit simple linear regression model
simple_model = sm.OLS(y, X_simple_const).fit()

# Calculate F-value and p-value for simple linear regression
fvalue_simple = simple_model.fvalue
pvalue_simple = simple_model.f_pvalue

# Prepare data for Ridge regression
X_ridge = correlation.drop(columns=['price'])  # Independent variables for Ridge regression

# Fit Ridge regression model
ridge_model = Ridge(alpha=1.0)  # You can adjust the regularization parameter alpha as needed
ridge_model.fit(X_ridge, y)

# Calculate predicted values for Ridge regression
y_pred_ridge = ridge_model.predict(X_ridge)

# Calculate residual sum of squares and total sum of squares for Ridge regression
rss_ridge = np.sum((y - y_pred_ridge) ** 2)
tss_ridge = np.sum((y - np.mean(y)) ** 2)

# Calculate degrees of freedom
n_obs = len(y)
k_ridge = len(X_ridge.columns)
df_residual_ridge = n_obs - k_ridge - 1
df_total_ridge = n_obs - 1

# Calculate F-value for Ridge regression
fvalue_ridge = ((tss_ridge - rss_ridge) / k_ridge) / (rss_ridge / df_residual_ridge)

# Calculate p-value for Ridge regression using scipy.stats.f
pvalue_ridge = 1 - fdist.cdf(fvalue_ridge, k_ridge, df_residual_ridge)

# Print results
print("Simple Linear Regression Results:")
print(simple_model.summary())
print(f"\nSimple Linear Regression F-value: {fvalue_simple:.4f}")
print(f"Simple Linear Regression p-value: {pvalue_simple:.4f}")

print("\nRidge Regression Coefficients:")
print(ridge_model.coef_)
print(f"\nRidge Regression F-value: {fvalue_ridge:.4f}")
print(f"Ridge Regression p-value: {pvalue_ridge:.4f}")
```

```python
#Summary of results
# Given correlation data
correlation_data = {
    'price': [1.000000, 0.308787, 0.525906, 0.701917, 0.089876, 0.256804, 0.053953, 0.036056],
    'bedrooms': [0.308787, 1.000000, 0.514508, 0.578212, 0.032471, 0.177944, 0.155670, 0.026496],
    'bathrooms': [0.525906, 0.514508, 1.000000, 0.755758, 0.088373, 0.502582, 0.507173, -0.126479],
    'sqft_living': [0.701917, 0.578212, 0.755758, 1.000000, 0.173453, 0.353953, 0.318152, -0.059445],
    'sqft_lot': [0.089876, 0.032471, 0.088373, 0.173453, 1.000000, -0.004814, 0.052946, -0.008830],
    'floors': [0.256804, 0.177944, 0.502582, 0.353953, -0.004814, 1.000000, 0.489193, -0.264075],
    'yr_built': [0.053953, 0.155670, 0.507173, 0.318152, 0.052946, 0.489193, 1.000000, -0.361592],
    'condition_coded': [0.036056, 0.026496, -0.126479, -0.059445, -0.008830, -0.264075, -0.361592, 1.000000]
}

# Convert correlation data to a DataFrame
df_correlation = pd.DataFrame(correlation_data)

# Prepare data
X = df_correlation.drop(columns=['price'])  # Independent variables
y = df_correlation['price']  # Dependent variable

# Fit Ridge regression models for each independent variable
results = {}
for col in X.columns:
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X[col].values.reshape(-1, 1), y, test_size=0.2, random_state=42)

    # Initialize and fit Ridge regression model
    ridge_model = Ridge(alpha=1.0)  # You can adjust the regularization parameter alpha as needed
    ridge_model.fit(X_train, y_train)

    # Evaluate model
    train_rmse = mean_squared_error(y_train, ridge_model.predict(X_train), squared=False)
    test_rmse = mean_squared_error(y_test, ridge_model.predict(X_test), squared=False)

    # Calculate F-value and p-value
    X_train_const = sm.add_constant(X_train)
    ridge_model_sm = sm.OLS(y_train, X_train_const).fit()
    fvalue = ridge_model_sm.fvalue
    pvalue = ridge_model_sm.f_pvalue

    # Store results
    results[col] = {
        'Coefficient': ridge_model.coef_[0],
        'Train RMSE': train_rmse,
        'Test RMSE': test_rmse,
        'F-value': fvalue,
        'P-value': pvalue
    }

# Print results
for col, result in results.items():
    print(f"Variable: {col}")
    print(f"Coefficient: {result['Coefficient']:.4f}")
    print(f"Train RMSE: {result['Train RMSE']:.4f}")
    print(f"Test RMSE: {result['Test RMSE']:.4f}")
    print(f"F-value: {result['F-value']:.4f}")
    print(f"P-value: {result['P-value']:.4f}")
    print()
```
