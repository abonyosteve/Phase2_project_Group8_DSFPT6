# D.) <span style='color:Red'>***Modelling***</span>

We are going to try and predict the price using some of the predictor variables, i.e., simple linear regression before advancing to the more multiple linear regression.

From our correlation results, we found that the top 3 most correlated variables to price are the `sqft_living`, `bathrooms` and `bedrooms` respectively; We are going to use these 3 to try and see how we can produce simple models and in the same way also see the results from remaining variables before merging all of them by multiple regression 

#### <span style='color:Green'>Simple linear regression</span>

```python
##importing the necessary library
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

## Let's choose the top 3 independent variables we are going to use in the order of correlation to price;
predictor_variables = ['sqft_living' , 'bathrooms', 'bedrooms', 'floors', 'sqft_lot', 'yr_built', 'condition_coded']

## Plotting a scatter plot of the above variables to price
for z in predictor_variables:
    plt.figure(figsize=(8,6))
    plt.scatter(df[z],df['price'], color = "blue")
    plt.title(f"Scatter plot of {z.capitalize()} and Price")
    plt.xlabel(z)
    plt.ylabel("Price")
    plt.grid(True)
    plt.show()
```

From the above results, the first 3 variables i.e., `sqft_living`, `bathrooms`, and `bedrooms` in that order, stand a chance of producing a better model compared to the  rest;  At this step, we assume that when we renovate these 3 features, the value of our homes will increase and therefore we will drop the other predictor variables.<br>
Let's go ahead and plot the line of best fit for these 3 variables using simple regression model

#### <span style='color:Green'>Simple linear regression Line of best fit</span>

```python
def line_bestfit(x,y,title,xlab,ylab):
    #let's get the mean of variable x and variable y
    meanX = np.mean(x)
    meanY = np.mean(y)
    
    #let's now calculate the slope
    #we are using the long method to calculate the slope
    # i.e., sum(x-mean(x) * y-mean(y)) / sum(*x-mean(x))squared)
    numerator = np.sum((x-meanX)*(y-meanY))
    denominator = np.sum((x-meanX)**2)
    slope = numerator/denominator
    
    #let's now calculate the y-intercept
    #from the formula; y = slope*x + c
    #the intercept will be c = meanY - slope*meanX
    y_intercept = ((meanY) - (slope*meanX))
    
    #printing the slope, the y-intercept, and the equation
    print(f"Slope: {slope}")
    print(f"Y-intercept: {y_intercept}")
    print(f"Equation of the line: y = {slope}X + {y_intercept}")    
    
    #plotting the original data and the line of best fit
    plt.figure(figsize=(8, 6))  #Set figure size
    plt.scatter(x, y ,color="blue", label="Data points") #original data points
    plt.plot(x, slope * x + y_intercept, color="red", label="Line of best fit") ## notice we've fixed the regression equation here
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.legend()
    plt.grid(True)
    plt.show()

##calling the function
print(line_bestfit(df['sqft_living'], df['price'], "Scatter plot of Sqft_living vs Price and the line of best fit", "sqft_living", "price"))
print(line_bestfit(df['bathrooms'], df['price'], "Scatter plot of Bathrooms vs Price and the line of best fit", "bathrooms", "price"))
print(line_bestfit(df['bedrooms'], df['price'], "Scatter plot of Bedrooms vs Price and the line of best fit", "bedrooms", "price"))
```

#### <span style='color:Green'>Simple linear regression using statmodels</span>

Now lets take our highly correlated variable `sqft_living` and build its model using stats model and proceed to interpret its results

```python
## importing the necessary libary
import statsmodels.api as sm

# Determining X and Y values
column_interest = ["sqft_living"]
X = df[column_interest]
y = df["price"]

# Create an OLS model      
model = sm.OLS(endog=y, exog=sm.add_constant(X))
# Get model results
results = model.fit()
# Display results summary
print(results.summary())
```

***Interpretation***

- The model and coefficients are statistically significant with all the p-values below 0.05
- R-squared is 49%; meaning only 49% of the variance in house prices can be explained by renovations in of the sqft_living
- The y-intercept(house price) is -43988.892194227315, meaning, if we did zero renovations on the sqft_living of the house,  we would still see a decrease of about -43988.892194227315
- The sqft_living coefficient is 280.8630, meaning, for each additional renovation in the sqft living; we would would see an improvement of 280 in the housevalue

Our model pointed out a case of strong multicollinearity; lets run again the correlation test to view this

#### <span style='color:Green'>Multiple linear regression using 2 independent variables</span>

```python
columns_interest = ["sqft_living", 'bathrooms']

## Determmining X and Y values
X_2 = df[columns_interest]
y = df["price"]

## creating OLS model
model2 = sm.OLS(y, sm.add_constant(X_2))

# Get model results
results2 = model2.fit()

# Display results summary
print(results2.summary())
```
